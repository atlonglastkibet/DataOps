# Matplotlib Example workflow


```python
#Import Matplotlib
%matplotlib inline

import matplotlib.pyplot as plt

#Import Data
x = [ 1, 3, 5, 8, 10]
y = [10, 25, 33, 50, 68]

#Setup plot
fig, ax = plt.subplots(figsize = (10,10)) #width and height

#Plot data
ax.plot(x,y)

#Customize plot
ax.set(title = 'Cars Plot',
       xlabel = 'Cars',
       ylabel = 'No. of cars'    
)
# Save and show figure
#fig.savefig('images/Cars-plot.png')

#show plot
plt.show()
```

# 2. Scatter plot

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate k values using linspace
k = np.linspace(1, 10, 100)  # 100 points between 1 and 10

# Setup plot
fig, ax = plt.subplots()

# Plot data (scatter plot)
ax.scatter(k, np.exp(k))  # Scatter plot of k vs exp(k)

# Show plot
plt.show()
```

# 3. Bar plots

```python
import matplotlib.pyplot as plt
%matplotlib inline

cars = {'Ferrari' : 20,
        'Ford' : 60,
        'Tesla' : 35
}

fig, ax = plt.subplots()

ax.bar(cars.keys(), cars.values())

ax.set(title = 'Car Count',
       xlabel = 'Make',
       ylabel = 'Count'
      )

plt.show();
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

cars = {'Ferrari' : 20,
        'Ford' : 60,
        'Tesla' : 35
}

fig, ax = plt.subplots()

ax.barh(cars.keys(), cars.values())

ax.set(title = 'Car Count',
       xlabel = 'Count',
       ylabel = 'Make'
      )

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 10))

ax1.hist(j)
ax1.set(title = "Histogram",
        xlabel = "Scores",
        ylabel = "Count")

ax2.plot(a,b)
ax2.set(title = "Line Graph",
        xlabel = "Car sales",
        ylabel = "Count")

ax3.bar(x,y)
ax3.set(title = "Bar Graph",
        xlabel = "Year",
        ylabel = "Sales")

ax4.barh(a,b)
ax4.set(title = "Horizontal Bar Graph",
        xlabel = "Sales",
        ylabel = "Year")
plt.show()
```

# What plot to use
1. When plotting something quick use pyplot
2. When plotting something advanced used matplotlib oo method

```python
heart_disease.age.dtype
```

```python
over_50 = heart_disease[heart_disease['age'] > 50] 
```

```python
over_50.head()
# over_50.drop(columns = 'over_50')
```

```python
# pyplot method
over_50.plot(kind = 'scatter',
             y = 'age',
             x = 'chol',
             c = 'target'
            )

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = 'scatter',
         x = 'age',
         y = 'chol',
         c = 'target',
            ax = ax
        )
ax.set_xlim([45,80])
plt.show()
```

## Object oriented method from scratch

```python
# 1. Create axes
fig, ax = plt.subplots(figsize = (10,6))

#2. Plot
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target']
)
# 3.Customise plot
ax.set(title = 'Heart disease data',
      xlabel = 'Age',
      ylabel = 'cholesterol')

#4. Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target')

#5. Add a horizontal line
ax.axhline(over_50['chol'].mean(), 
          linestyle =  '--')


plt.show()
```

## Subplots

```python
fig, ((ax0, ax1)) = plt.subplots(nrows = 1,
                                 ncols = 2,
                                 figsize = (10, 5))

ax0.scatter(x = over_50['age'],
            y = over_50['chol'],
            c = over_50['target'],
            cmap = 'winter'
        )
ax0.set(title = 'Cholesterol',
        xlabel = 'Age',
        ylabel = 'Cholesteral')

ax1.scatter(x = over_50['age'],
            y = over_50['thalach'],
            c = over_50['target'],
           cmap = 'coolwarm_r')
ax1.set(title = 'Heart beat',
        xlabel = 'Age',
        ylabel = 'Heat rate')

fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold')

plt.show();
```
