## 1. Getting the data ready

- Split data into features and labels
- Filling or disregard missing samples

### a. Standard imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import pandas and read from our data
import pandas as pd
heart_df = pd.read_csv('92 - heart-disease.csv')
heart_df.head()
```

### b. Create x and y variables

- Use all the columns `'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal'` to predict the `'target'` (1 or 0) to mean heart disease or not. 

- x variable will be the features matrix, y will be the target

```python
# Features
x = heart_df.drop('target', axis = 1)
```

```python
# Labels
y = heart_df['target']
```

## 2. Choose the right estimator/algorithms for our problems

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# Keep the default hyperparameters

clf.get_params()
```

## 3. Fit the model/algorithm and use it to make the right prediction on our data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2 )
```

```python
clf.fit(x_train, y_train)
```

```python
type(x_train)
```

```python
type(y_train)
```

### b) Make a prediction

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
y_preds = clf.predict(x_test)
y_preds
```

## 4. Evaluate the model

```python
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

## 5. Improve a model

```python
import numpy as np
np.random.seed(42)

for i in range(10, 100, 10):
    print(f'Model fitting with {i} estimators...')
    clf = RandomForestClassifier().fit(x_train, y_train)
    print(f'Model accuracy: {clf.score(x_test, y_test)*100:.2f}%')
    print(' ')
```

## 6. Save the model and load it

```python
import pickle
pickle.dump(clf, open('heart-disease-random-forest-model.pkl', 'wb'))
```

```python
loaded_model = pickle.load(open('heart-disease-random-forest-model.pkl', 'rb'))

loaded_model.score(x_test, y_test)
```
